```python
url = 'https://data.nasdaq.com/api/v3/datatables/ZILLOW/DATA?indicator_id=ZSFH&region_id=394463&api_key=nhngZKzAkdnohAMb46Kx'
response = requests.get(url)
data = response.json()
```

```python
if response.status_code == 200:
    # Convert response to JSON format
    data = response.json()
    
    # Extract relevant data
    zillow_data = data['datatable']['data']
    column_names = [column['name'] for column in data['datatable']['columns']]
    
    # Create DataFrame
    zillow_df = pd.DataFrame(zillow_data, columns=column_names)
```

```python
zillow_df['date'] = pd.to_datetime(zillow_df['date'])
zillow_df
```

```python
dates = zillow_df['date']
prices = zillow_df['value']

plt.plot(dates, prices)
plt.title('Chicago House Prices of Single Family Homes')
plt.xlabel('Date')
plt.ylabel('Price ($)')  
plt.tight_layout()

# Display the plot
plt.show()
```
