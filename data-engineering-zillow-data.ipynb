```python
import pandas as pd
import numpy as np
import re
```

```python
zillow_data = pd.read_csv('Resources/Zillow_Data.csv')
zillow_data.head()
```

```python
zillow_indicators = pd.read_csv('Resources/Zillow_Indicators.csv')
zillow_indicators.head()
```

```python
zillow_regions = pd.read_csv('Resources/Zillow_Regions.csv')
zillow_regions.head()
```

```python
merge_one_df = pd.merge(zillow_data,zillow_indicators,on='indicator_id')
merged_df = pd.merge(merge_one_df,zillow_regions,on='region_id')
merged_df
```

```python
p = 'Chicago'
chicago_df = merged_df[merged_df['region'].str.contains(p)]
chicago_df.tail()
print(len(chicago_df))
```
